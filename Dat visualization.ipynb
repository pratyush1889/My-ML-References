```python
# importing basic libraries for data reading and visualization
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Reading the data from a csv file
data_set = pd.read_csv("train.csv")
data_set.head(3)
```

```python
#combining two derived columns into one
data_set["Family"] = data_set["SibSp"] + data_set["Parch"]
data_set.head(3)
```

```python
# describing data to undertsnad the count, mena, standard deviation
data_set.describe()
```

```python
# gives basic info about the dataset
data_set.info()
```

```python
# describes only categorical data
data_set.describe(include=['O'])
```

```python
# describes  all features in the data set
data_set.describe(include='all')
```

```python
# tells you the correlation between different features in the dats set
data_set.corr()
```

##### Correlation is a measure of joint variablity of two variables
A correlation of 1 signifys perfect positive relationship, meaning one varable is perfectly explained by other variable
A correlation of 0 implies there is no relationship betwenn two variables and are independent
A corrleation of -1 implies that there is  a perfect negative relationship between two variables

Based on the above table we make following inferences

1. Passenger Id does not have a significant impact on Passenger's Survival
2. Pclass has a negative correlationship with Survival means a third class passenger has a low survival chance
3. Age has a negative correlationship with Survival. Higher the age lower the survival chance
4. Sibblings and Spouses seems to be have a negative correlation with Suvival chance
5. Parents and Child have a postive  corelation with Survival chance
6. when we combine spouse, sibbling, parents and child as Family, we see a slight postive correlation with Survival chance

```python
# understanding the male and female data distribution
# Visulaizing ratio of males and females
plt.figure(figsize=(3,3))
sns.countplot(data=data_set, x="Sex")
```

```python
# understanding the data age distribution
sns.distplot(data_set["Age"].dropna(), bins=np.linspace(0,100,10))
```

```python
# understanding the How many family memebers distribution
sns.factorplot(data=data_set, x="Family", kind="count")
```

```python
# Violinplot shows the kernel density estimation
# Bandwidth of the kernel shows the influence on the  resulting estimate
# fatter  bandwidth implies more males did not survive however fatter bandwidth for females shows that more females survied
sns.violinplot(x= "Sex", y ="Survived", data=data_set)
```

```python
# What was the ratio of Males and Females in different Class
sns.factorplot(data = data_set, x="Pclass", hue="Sex", kind="count")
```

```python
# understanding various features influence on passengers's survival
# How sex of the pasenger impacted their survival
sns.factorplot(data = data_set, x="Sex", hue="Survived", kind="count")
```

```python
# using facetgrid
# Facet Grid helps to visulalize the distribution of variable or 
# relationship between multiple variable within a subsets of your dataset
# like Pclass and Survival, where we have Pclass 1,2 and 3 and within each class how many passenges survived
pclass = sns.FacetGrid(data_set, row="Pclass", col="Survived", size=3, aspect=2)
pclass.map(plt.hist,'Age',alpha=0.5, bins=10)
pclass.add_legend()
```
