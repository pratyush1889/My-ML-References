##### Correlation is a measure of joint variablity of two variables
A correlation of 1 signifies perfect positive relationship, meaning one variable is perfectly explained by other variable
A correlation of 0 implies there is no relationship betwenn two variables and are independent
A corrleation of -1 implies that there is  a perfect negative relationship between two variables

Based on the above table we make following inferences

1. Passenger Id does not have a significant impact on Passenger's Survival
2. Pclass has a negative correlationship with Survival means a third class passenger has a low survival chance
3. Age has a negative correlationship with Survival. Higher the age lower the survival chance
4. Sibblings and Spouses seems to be have a negative correlation with Suvival chance
5. Parents and Child have a postive  corelation with Survival chance
6. when we combine spouse, sibbling, parents and child as Family, we see a slight postive correlation with Survival chance
